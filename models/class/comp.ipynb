```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from joblib import load
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
X_train, X_test, y_train, y_test = load('../../preprocess/classification_data.joblib')
model_names2,accuracy_train2 ,precision_train2 ,recall_train2 ,f1_train2 ,accuracy_test2 ,precision_test2 ,recall_test2 ,f1_test2= load('tree.joblib')
model_names,accuracy_train ,precision_train ,recall_train ,f1_train ,accuracy_test ,precision_test ,recall_test ,f1_test = load('knn.joblib')
```

```python
model_names+=model_names2
accuracy_train+=accuracy_train2
precision_train+=precision_train2
recall_train+=recall_train2
f1_train+=f1_train2
accuracy_test+=accuracy_test2
precision_test+=precision_test2
recall_test+=recall_test2
f1_test+=f1_test2
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(35, 20))

axes[0, 0].bar(model_names, accuracy_train)
axes[0, 0].set_title('Accuracy')
axes[0, 0].set_ylim(0, 1)

axes[0, 1].bar(model_names, precision_train)
axes[0, 1].set_title('Precision')
axes[0, 1].set_ylim(0, 1)

axes[1, 0].bar(model_names, recall_train)
axes[1, 0].set_title('Recall')
axes[1, 0].set_ylim(0, 1)

axes[1, 1].bar(model_names, f1_train)
axes[1, 1].set_title('F1 score')
axes[1, 1].set_ylim(0, 1)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(35, 20))

axes[0, 0].bar(model_names, accuracy_test)
axes[0, 0].set_title('Accuracy')
axes[0, 0].set_ylim(0, 1)

axes[0, 1].bar(model_names, precision_test)
axes[0, 1].set_title('Precision')
axes[0, 1].set_ylim(0, 1)

axes[1, 0].bar(model_names, recall_test)
axes[1, 0].set_title('Recall')
axes[1, 0].set_ylim(0, 1,)

axes[1, 1].bar(model_names, f1_test)
axes[1, 1].set_title('F1 score')
axes[1, 1].set_ylim(0, 1)
```
